```python
def fix_time(x):
    # If x is already a pandas Timestamp, convert it directly.
    if isinstance(x, pd.Timestamp):
        return "0 days " + x.strftime("%H:%M:%S")
    s = str(x)
    # Check if it's already in timedelta string form
    if s.startswith('0 days'):
        return s
    # If it contains a colon, assume it's a time string
    if ":" in s:
        return "0 days " + s
    # Otherwise, try to parse it as a date and extract its time part
    try:
        dt = pd.to_datetime(s, errors='coerce')
        if pd.notnull(dt):
            return "0 days " + dt.strftime("%H:%M:%S")
    except Exception:
        pass
    return "0 days 00:00:00"

# Fix the 'Time' column values
dec_2024_df['Time'] = dec_2024_df['Time'].apply(fix_time)

# Convert 'Time' column to timedelta
dec_2024_df['Time'] = pd.to_timedelta(dec_2024_df['Time'], errors='coerce')

# Combine 'Date' and 'Time' into a single datetime column
dec_2024_df['DateTime'] = dec_2024_df['Date'] + dec_2024_df['Time']

# Sort by bike and datetime
dec_2024_df.sort_values(by=['Bike Number', 'DateTime'], inplace=True)

# Build a list of custody records
records = []
for bike_num, group in dec_2024_df.groupby('Bike Number'):
    group = group.reset_index(drop=True)
    current_rider = None
    current_start = None

    for _, row in group.iterrows():
        if row['IN & Out Remarks'] == 'Received':
            if current_rider is not None:
                records.append({
                    'Bike Number': bike_num,
                    'Rider Name': row.get('Rider Name'),
                    'Rider ID': current_rider,
                    'Issued At': current_start,
                    'Received At': row['DateTime'],
                    'working platform': row['working platform'],
                    'Vendors': row['Vendors']
                })
                current_rider = None
                current_start = None
        elif row['IN & Out Remarks'] == 'Issue':
            if current_rider is not None:
                # Close previous custody if it wasn't received
                records.append({
                    'Bike Number': bike_num,
                    'Rider Name': row.get('Rider Name'),
                    'Rider ID': current_rider,
                    'Issued At': current_start,
                    'Received At': row['DateTime'],
                    'working platform': row['working platform'],
                    'Vendors': row['Vendors']
                })
            # Start a new custody record
            current_rider = row['Rider ID']
            current_start = row['DateTime']

    # If a bike remained issued at the end
    if current_rider is not None:
        records.append({
            'Bike Number': bike_num,
            'Rider ID': current_rider,
            'Issued At': current_start,
            'Received At': None,
            'working platform': row['working platform'],
            'Vendors': row['Vendors']
        })

custody_df = pd.DataFrame(records)

# Ensure 'Issued At' and 'Received At' columns have proper datetime values
custody_df['Issued At'] = pd.to_datetime(custody_df['Issued At'], errors='coerce')
custody_df['Received At'] = pd.to_datetime(custody_df['Received At'], errors='coerce')

custody_df
```
