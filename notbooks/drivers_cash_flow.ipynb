# Load and Extract Data

# Save all data into an excel file

```python
date_time = datetime.now().strftime('%m%d%Y')
date_time
```

```python
folder_name = '../data/processed/'
os.makedirs(folder_name, exist_ok=True)

df_pivot.to_excel(f'{folder_name}deliveroo_{date_time}.xlsx', index=False)
print('File saved')
```
